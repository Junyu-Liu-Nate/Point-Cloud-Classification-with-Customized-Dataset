```python
import numpy as np
import trimesh
from plyfile import PlyData, PlyElement

def convert_off_to_ply(off_file_path, ply_file_path):
    # Read .off file using trimesh
    mesh = trimesh.load(off_file_path)

    # Extract vertices and faces
    vertices = mesh.vertices
    # print(vertices)
    faces = mesh.faces
    # print(faces)

    # Convert vertices to the format PlyElement expects
    vertex_elements = []
    for vertex in vertices:
      vertex_elements.append((vertex[0], vertex[1], vertex[2]))
    vertex_dtype = [('x', 'f4'), ('y', 'f4'), ('z', 'f4')]
    vertex_elements = np.array(vertex_elements, dtype=vertex_dtype)

    # Convert faces to the format PlyElement expects
    face_indices = []
    faces = faces.tolist()
    print(faces)
    # for face in faces:
    #   face_indices.append(([face[0], face[1], face[2]]))
    face_indices = [tuple(inner_list) for inner_list in faces]
    print(face_indices)
    face_dtype = [('vertex_indices', 'i4', (3,))]
    face_indices = np.array(face_indices, dtype=face_dtype)
    print(face_indices)

    # Create PlyData object
    vertex_element = PlyElement.describe(vertex_elements, 'vertex')
    face_element = PlyElement.describe(face_indices, 'face')
    ply_data = PlyData([vertex_element, face_element])

    # Write .ply file
    with open(ply_file_path, 'wb') as ply_file:
        ply_data.write(ply_file)
```

```python
# Get a list of all subdirectories (subfolders) within the main folder
subfolder_names = [name for name in os.listdir(projectPath + dataPath) if os.path.isdir(os.path.join(projectPath + dataPath, name))]

for subfolder_name in subfolder_names:
  classList.append(subfolder_name)

  trainvalFilePath = subfolder_name + '/train/'
  testFilePath = subfolder_name + '/test/'

  trainFileNames = [name for name in os.listdir(trainvalFilePath) if name.lower().endswith('.off')]
  testFileNames = [name for name in os.listdir(testFilePath) if name.lower().endswith('.off')]

  for trainFileName in trainFileNames:
    offFilePath = subfolder_name + '/train/' + trainFileName
    plyFilePath = offFilePath[:-3] + 'ply'
    convert_off_to_ply(offFilePath, plyFilePath)

  for testFileName in testFileNames:
    offFilePath = subfolder_name + '/test/' + testFileName
    plyFilePath = offFilePath[:-3] + 'ply'
    convert_off_to_ply(offFilePath, plyFilePath)
```
